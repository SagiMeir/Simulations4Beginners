```python
avgQE =[]
avgU =[]
amountbeads =[]

for i in range(2,8):
    bead = 2**i
    df = pd.read_csv("sim"+str(bead)+".log",sep='\s+')
    # plt.plot(df["step"].values[50:],df["QE"].values[50:])
    avgQE.append(df["QE"].values[50:].mean())
    avgU.append(df["U"].values[50:].mean()/bead*2)
    amountbeads.append(bead)
plt.plot(amountbeads, avgQE)
plt.plot(amountbeads, avgU, color = "red")
plt.axhline(4.005263763420102e-21)


plt.xlabel("amount of beads")
plt.ylabel("average estimator")
```
