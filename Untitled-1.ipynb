## PIMD const energy

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# and for nice looking plots:
import seaborn as sns
sns.set_theme(style="ticks",rc={'axes.labelsize':14},palette='tab10')

df = pd.read_csv("sim.log",sep='\s+')
x = df["step"].values
plt.plot(x,df["K"].values,label ="K")
plt.plot(x,df["U"].values,label ="U")
plt.plot(x,df["SpringEnergy"].values,label ="SpringEnergy")
plt.plot(x,df["E"].values,label ="TOTAL")
plt.legend()
```

## PIMD const temp

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# and for nice looking plots:
import seaborn as sns
sns.set_theme(style="ticks",rc={'axes.labelsize':14},palette='tab10')

df = pd.read_csv("sim.log",sep='\s+')
x = df["step"].values[100:]
plt.plot(x,df["T"][100:].values,label ="temp")
print((df["T"][100:].values).mean())
plt.legend()
```
