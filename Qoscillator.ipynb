```python
from typing import Optional
import numpy as np
import pandas as pd
from scipy.constants import Boltzmann as BOLTZMANN
from scipy.constants import hbar
import matplotlib.pyplot as plt
# and for nice looking plots:
import seaborn as sns
sns.set_theme(style="ticks",rc={'axes.labelsize':14},palette='tab10')
```

```python
omega =7.596E13 # 1/sec
Amp = 5 * 10**(-10) # meter
mass = 6.633E-26 # kg
temp = 96.7 # K

bead_omega =np.sqrt(64)*BOLTZMANN*temp/hbar
print(f"1 / omega p = {(1/bead_omega) }")
sys_temp = hbar*omega/BOLTZMANN/6
print(f"system temp = {sys_temp} K")
exp_estimator = hbar*omega*0.5 # J
print(f"the quantom energy = {exp_estimator} J")

U_theory = mass * omega**2 * (Amp)**2 /4
print(f"theoretical potential energy = {U_theory} J")


```
