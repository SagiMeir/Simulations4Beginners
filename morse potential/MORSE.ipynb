the tylor expention for morse potential
${\displaystyle V(r)\approx {\frac {1}{2}}k(r-r_{e})^{2}}$  || ${\displaystyle \,k=2D_{e}a^{2} = 2m\omega^{2}}$

$De\cdot \:2\left(1-e^{-a\left(x-Re\right)}\right)ae^{-a\left(x-Re\right)}$
